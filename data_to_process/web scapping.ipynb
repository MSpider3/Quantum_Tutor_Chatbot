```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_qiskit_article(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Failed to fetch {url}")
        return ""

    soup = BeautifulSoup(response.text, 'html.parser')

    # Try h1, fallback to h2 or URL
    title_tag = soup.find('h1') or soup.find('h2')
    title = title_tag.get_text(strip=True) if title_tag else url

    paragraphs = soup.find_all('p')
    content = "\n\n".join([p.get_text(strip=True) for p in paragraphs])

    markdown_text = f"# {title}\n\n{content}\n"
    return markdown_text
```

```python
urls = [
    "https://www.ibm.com/quantum/blog/iqp-upgrade",
    "https://www.ibm.com/quantum/blog/optimization-white-paper",
    "https://www.ibm.com/quantum/blog/japanese-quantum-ecosystem",
    "https://www.ibm.com/quantum/blog/supercomputing-24",
    "https://www.ibm.com/quantum/blog/qdc-2024",
    "https://www.ibm.com/think/topics/quantum-computing",
    "https://www.spinquanta.com/news-detail/what-is-quantum-computing-and-how-does-it-work-expert-explained",
    "https://www.spinquanta.com/news-detail/what-is-a-qubit",
    "https://learning.quantum.ibm.com/course/quantum-computing-in-practice/introduction",
    "https://learning.quantum.ibm.com/course/quantum-computing-in-practice/running-quantum-circuits",
    "https://learning.quantum.ibm.com/course/quantum-computing-in-practice/utility-scale-qaoa",
    "https://learning.quantum.ibm.com/course/quantum-computing-in-practice/what-problems-are-quantum-computers-good-for"
]

markdown_articles = []
for url in urls:
    article_md = scrape_qiskit_article(url)
    if article_md:
        markdown_articles.append(article_md)
        
import os

# Save the markdown articles to the file
os.makedirs("../data_to_process/markdown", exist_ok=True)
with open("../data_to_process/markdown/qiskit_articles.md", "w", encoding="utf-8") as md_file:
    for article in markdown_articles:
        md_file.write(article + "\n\n---\n\n")
print("Articles have been scraped and saved to qiskit_articles.md")
```
